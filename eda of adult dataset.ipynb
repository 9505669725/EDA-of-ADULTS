# outliers 

```python
out=[]
def Zscore_outlier(data):
    m=np.mean(data)
    sd=np.std(data)
    for i in data:
        z=(i-m)/sd
        if np.abs(z)>3:
            out.append(i)
    print("outliers:",out)
```

```python
Zscore_outlier(data['age'])
```

```python
Zscore_outlier(data['educational-num'])
```

```python
Zscore_outlier(data['capital-loss'])
```

```python
Zscore_outlier(data['capital-gain'])
```

# find IQR 

```python
out=[]
def iqr_outlier(data):
    q1=data.quantile(0.25)
    q3=data.quantile(0.75)
    iqr=q3-q1
    lower=q1-1.5*iqr
    upper=q3+1.5*iqr
    for i in data:
        if i >upper or i< lower:
            out.append(i)
    print("outliers:",out)
```

```python
iqr_outlier(data['age'])
```

```python
iqr_outlier(data['hours-per-week'])
```

# imputation

```python
sns.boxplot(data['age'])
plt.title('boxplot before mean impution ')
plt.show()
q1=data['age'].quantile(0.25)
q3=data['age'].quantile(0.75)
iqr=q3-q1
lower=q1-1.5*iqr
upper=q3+1.5*iqr
m=np.mean(data['age'])
for i in data['age']:
    if i > upper or i < lower:
        data['age']=data['age'].replace(i,m)
sns.boxplot(data['age'])
plt.title('boxplot after mean impution ')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
data_num=pd.DataFrame(scale.fit_transform(numerical),columns=numerical.columns)
```

```python
categorical=data.select_dtypes(include=['object'])
```

```python
categorical.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
enc=OneHotEncoder(sparse=False)
data_cat=pd.DataFrame(enc.fit_transform(categorical),columns=enc.get_feature_names(categorical.columns))
```

```python
y=data.pop('income_>50K')
X=data
```

```python
X.pop('income_<=50K')
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
lg.fit(X_train,y_train)
```

```python
y_pred=lg.predict(X_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test,y_pred)
```

```python
metrics.plot_confusion_matrix(lg,X_test,y_test)
```

```python
print(metrics.classification_report(y_test,y_pred))
```
